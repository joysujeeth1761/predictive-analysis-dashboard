i found that : the target attribute G3 has a strong correlation with attributes G2 and G1. This occurs because G3 is the final year grade (issued at the 3rd period), while G1 and G2 correspond to the 1st and 2nd-period grades. It is more difficult to predict G3 without G2 and G1, but such prediction is much more useful

```python
X = df.drop(columns=["G1","G2","G3"])
y = df['G3']
```

```python
corr = df.corr(numeric_only=True)["G3"].sort_values(ascending=False)

print(corr)
```

EDA Findings

1. G1 and G2 show extremely high correlations with G3 (0.826 and 0.919), indicating potential target leakage.

2. Previous academic failures show the strongest negative correlation (-0.393) among behavioral features.

3. Study time is positively associated with academic performance (0.25).

4. Higher parental education levels correlate with better student outcomes.

5. Alcohol consumption exhibits a negative relationship with final grades.

6. Travel time and absences have mild negative impacts on performance.

```python
#Feature engineering 
#1
df["total_alcohol"] = df["Dalc"]+df["Walc"]
#dalc  = daily alcohol
#Wlac = weekend alcohol
```

```python
#2 parent education score

df["parent_edu"] = df['Medu']+df['Fedu']

```

```python
# study efficiency
df["study_failure_ratio"] = df["studytime"] /df['failures'] + 1

#Student studying 4 hours with 0 failures ≠ student studying 4 hours with 4 failures.
```
